### Adding New Functions

- We can create our own functions but they are local and not global; they stay within the script or file within we create them
- They only work after they are called 
- To create a function you have to declare the function, then give it a name (this is called the header). We declare a function with the def keyword and give it a name
    - When naming a function, you cannot use a number as the starting point
    - You cannot name is with a keyword 
    - After the name you need to include a bracket (), and you may declare a variable or not declare a variable in the bracket 
    - Afterwards you need to end the naming with a colon



```python
def print_lyrics():
    print("I am a good boy!")
    print("I am a good boy that is learning how to code!")
```

### Flow of Execution

- When I declare and I call a function python will read the statements in the function from top to bottom

- But when I call another function inside the function, it runs from top and when it gets to that function, it executes the statements in that function, and then come back from where it left off in the original function and then complete the function is that it? 

## Parameters and Arguments:

- **Parameters** are the declarations of the variables you make when creating your function

- **Arguments** are the inputs we make into our functions after it has been created and when we are calling the function.

- A function may or maynot take in arguments 

### Fruitful Functions vs Void Functions

- Void functions do not return a value, they can print or not do anything
- Fruitful functions returns a value you can use later on
- A function may or may not return a value
